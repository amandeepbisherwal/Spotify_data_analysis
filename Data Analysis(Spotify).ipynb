### 2. top 10 songs by popularity

```python
spotify_top_track_names = spotify_tracks.groupby('track_name')['popularity'].sum().sort_values(ascending = False).head(10)
spotify_top_track_names
```

```python
plt.figure(figsize = (15,6))
spotify_top_track_names.plot.bar(color = 'mediumseagreen')
plt.legend()
plt.title('Top 10 songs by popularity', size = 15)
plt.xlabel('Songs')
plt.ylabel('Popularity count')
plt.show()
```

### 4. Most stream artists

```python
top_chart_artist = spotify_artist.groupby('name')['popularity'].count().sort_values(ascending=False).head()
top_chart_artist
```

```python
plt.figure(figsize = (8,5))
plt.title('Most stream artist')
top_chart_artist.plot.bar(color  = 'slategray')
plt.xlabel('Artists')
plt.ylabel('Popularity count')
plt.show()
```

#### Rico, Ace, Forr√≤ Cintura de Mola are the most stream artists.

### Artist with having highest followers on Spotify.

```python
artist_follower = spotify_artist.groupby('name')['followers'].sum().sort_values(ascending = False).head(15)
artist_follower
```

```python
plt.figure(figsize = (15,9))
plt.title('Most followed artist', size = 15)
artist_follower.plot.bar(color = 'forestgreen')
plt.xlabel('artists')
plt.ylabel('followers count(in millions)')
plt.show()
```

#### Ed Sheeran have the highest followers count among the top 15 top artist.
